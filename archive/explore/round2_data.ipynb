```python
def dist(x1, y1, x2, y2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)
```

```python
change1 = 0
non_change1 = 0
change2 = 0
non_change2 = 0
    
for sid in range(len(sequence_names)):
    keypoints = sequences[sequence_names[sid]]['keypoints']
    last_x_1, last_y_1, last_x_2, last_y_2 = keypoints[0]
    for (x1, y1, x2, y2) in keypoints:
        if dist(x1, y1, last_x_1, last_y_1) > 0.0000001:
            change1 += 1
        else:
            non_change1 += 1
            
        if dist(x2, y2, last_x_2, last_y_2) > 0.0000001:
            change2 += 1
        else:
            non_change2 += 1
            
        last_x_1, last_y_1, last_x_2, last_y_2 = x1, y1, x2, y2

print(f"change1 / non_change1 = {change1} / {non_change1}")
print(f"change2 / non_change2 = {change2} / {non_change2}")
```

## Moving label

```python
import os
import numpy as np
datafolder = './data'
keypoints = np.load(os.path.join(datafolder, 'submission_keypoints.npy'), allow_pickle=True).item()
```

```python
sequences = keypoints['sequences']
sequence_names = sequences.keys()
sequence_names = [sequence_name for sequence_name in sequence_names]
print(f"Number of sequences: {len(sequence_names)}")

sequence0_keypoints = sequences[sequence_names[0]]['keypoints']

print(f"Keypoinys shape: {sequence0_keypoints.shape}")
```

```python
def dist(x1, y1, x2, y2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)
```

```python
moving_label = {}

for sid in range(len(sequence_names)):
    label = []
    keypoints = sequences[sequence_names[sid]]['keypoints']
    last_x_1, last_y_1, last_x_2, last_y_2 = keypoints[0]
    for (x1, y1, x2, y2) in keypoints:
        state1, state2 = 0, 0 
        if dist(x1, y1, last_x_1, last_y_1) > 0.0000001:
            state1 = 1
        else:
            state1 = 0
            
        if dist(x2, y2, last_x_2, last_y_2) > 0.0000001:
            state2 = 1
        else:
            state2 = 0
            
        label.append([state1, state2])
    
    moving_label[sequence_names[sid]] = np.array(label)
    
    if sid % 1000 == 0:
        print(f"Processing {sid}")

np.save('./data/moving_labels.npy', moving_label, allow_pickle=True)
```
